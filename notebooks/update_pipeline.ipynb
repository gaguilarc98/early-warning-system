```python
from matplotlib.figure import Figure

BG = "#0e1117"
TEXT = "#e8eaf0"
SPINE = "#2d3140"

ALERT_COLORS = {
    "None":   "#2a4858",
    "Orange": "#ff8c00",
    "Red":    "#e63946",
}

ALERT_ORDER = ["None", "Orange", "Red"]


def _apply_dark_style(fig: Figure, axes) -> None:
    fig.patch.set_facecolor(BG)
    for ax in (axes if hasattr(axes, "__iter__") else [axes]):
        ax.set_facecolor(BG)
        ax.tick_params(colors=TEXT, labelsize=8)
        ax.xaxis.label.set_color(TEXT)
        ax.yaxis.label.set_color(TEXT)
        ax.title.set_color(TEXT)
        for spine in ax.spines.values():
            spine.set_edgecolor(SPINE)
            
def plot_probability(
    lon_grid: np.ndarray,
    lat_grid: np.ndarray,
    prob_grid: np.ndarray,
    date: str = "",
    figsize: tuple[int, int] = (8, 6),
) -> Figure:
    """
    Pcolormesh of continuous predict_prob values.

    Parameters
    ----------
    lon_grid, lat_grid : 2-D meshgrid arrays
    prob_grid          : 2-D array of probabilities [0, 1]
    date               : string label for the title
    """
    fig, ax = plt.subplots(figsize=figsize)
    _apply_dark_style(fig, ax)

    cmap = plt.cm.RdYlBu_r
    norm = mcolors.Normalize(vmin=0, vmax=1)

    pcm = ax.pcolormesh(lon_grid, lat_grid, prob_grid,
                        cmap=cmap, norm=norm, shading="auto")

    cbar = fig.colorbar(pcm, ax=ax, fraction=0.03, pad=0.02)
    cbar.set_label("P(coldspell)", color=TEXT, fontsize=10)
    cbar.ax.yaxis.set_tick_params(color=TEXT, labelsize=8)
    plt.setp(cbar.ax.yaxis.get_ticklabels(), color=TEXT)

    # Threshold contour lines
    for label, thresh in [("Orange", 0.33), ("Red", 0.66)]:
        cs = ax.contour(lon_grid, lat_grid, prob_grid,
                        levels=[thresh],
                        colors=[ALERT_COLORS[label]],
                        linewidths=1.0, linestyles="--")
        ax.clabel(cs, fmt={thresh: f"{label} ({thresh:.0%})"},
                  colors=[ALERT_COLORS[label]], fontsize=7)

    ax.set_aspect("equal")
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    title = "Coldspell probability"
    if date:
        title += f"  ·  {date}"
    ax.set_title(title, pad=10, fontsize=11)

    fig.tight_layout()
    return fig
```

```python
def forecast_to_grid(
    forecast_df: pd.DataFrame,
    value_col: str,
) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Pivot a flat lon/lat DataFrame column into a 2-D grid for pcolormesh.

    Returns
    -------
    lon_grid : 2-D array
    lat_grid : 2-D array
    values   : 2-D array of `value_col`
    """
    lons = np.sort(forecast_df["lon"].unique())
    lats = np.sort(forecast_df["lat"].unique())

    pivot = forecast_df.pivot(index="lat", columns="lon", values=value_col)
    pivot = pivot.reindex(index=lats, columns=lons)

    lon_grid, lat_grid = np.meshgrid(lons, lats)
    return lon_grid, lat_grid, pivot.values
```

```python
import matplotlib.patches as mpatches
def plot_alert_level(
    lon_grid: np.ndarray,
    lat_grid: np.ndarray,
    alert_grid: np.ndarray,
    date: str = "",
    figsize: tuple[int, int] = (8, 6),
) -> Figure:
    """
    Pcolormesh of categorical alert levels (None / Orange / Red).

    Parameters
    ----------
    lon_grid, lat_grid : 2-D meshgrid arrays
    alert_grid         : 2-D array of strings ('None', 'Orange', 'Red')
    date               : string label for the title
    """
    # Encode categories to integers for pcolormesh
    level_to_int = {lvl: i for i, lvl in enumerate(ALERT_ORDER)}
    int_grid = np.vectorize(level_to_int.get)(alert_grid).astype(float)

    cmap = mcolors.ListedColormap([ALERT_COLORS[l] for l in ALERT_ORDER])
    norm = mcolors.BoundaryNorm(boundaries=[-0.5, 0.5, 1.5, 2.5], ncolors=3)

    fig, ax = plt.subplots(figsize=figsize)
    _apply_dark_style(fig, ax)

    ax.pcolormesh(lon_grid, lat_grid, int_grid,
                  cmap=cmap, norm=norm, shading="auto")

    # Manual legend
    patches = [
        mpatches.Patch(color=ALERT_COLORS[lvl], label=lvl)
        for lvl in ALERT_ORDER
    ]
    legend = ax.legend(
        handles=patches, loc="lower right", framealpha=0.25,
        facecolor=BG, edgecolor=SPINE, labelcolor=TEXT, fontsize=9,
    )

    ax.set_aspect("equal")
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    title = "Alert level"
    if date:
        title += f"  ·  {date}"
    ax.set_title(title, pad=10, fontsize=11)

    fig.tight_layout()
    return fig
```

```python
plot_alert_level(lon_grid, lat_grid, alert_grid, date='2026-01-01')
```
